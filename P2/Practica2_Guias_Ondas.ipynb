## 2. Importación de datos

Para importar datos en celdas de código Python, se pueden utilizar diferentes métodos dependiendo de la fuente de los datos:

### 2.1 Anotando directamente los valores en el código
Se pueden definir listas o diccionarios directamente en el código para representar los datos experimentales. Por ejemplo:

````python
# Datos experimentales anotados directamente
frecuencias = [1e6, 2e6, 4e6]  # Frecuencias en Hz
desfases = [0.1, 0.2, 0.4]     # Desfases en radianes

print(f"Frecuencias: {frecuencias}")
print(f"Desfases: {desfases}")
````

### 2.2 Importando datos desde un fichero CSV
Se pueden leer tablas de datos experimentales almacenadas en un archivo CSV utilizando bibliotecas como `pandas`.

Por ejemplo, un archivo llamado `datos.csv` podría tener el siguiente contenido:

```
Frecuencia,Desfase
1e6,0.1
2e6,0.2
4e6,0.4
```

Este fichero se podría importar a un bloque de código Python. Por ejemplo:

````python
import pandas as pd

# Leer datos desde un archivo CSV
datos_df = pd.read_csv('datos.csv')

# Mostrar los primeros registros
print(datos_df.head())
````

Ambos métodos son útiles dependiendo de la cantidad y complejidad de los datos a analizar.

## 3. Manipulación de datos

Para manipular los datos previamente importados, se pueden realizar diversas operaciones como filtrado, transformación, o cálculo de estadísticas. Por ejemplo, utilizando `pandas`, es posible seleccionar columnas específicas, aplicar funciones matemáticas, o filtrar filas según condiciones. Esto permite analizar y preparar los datos para su visualización o modelado. 

Un ejemplo básico sería calcular el promedio de la columna `Desfase` del DataFrame `datos_df`:

````python
promedio_desfase = datos_df['Desfase'].mean()
print(f"Promedio del desfase: {promedio_desfase}")
````

Este ejemplo muestra cómo utilizar la biblioteca `numpy` para realizar operaciones matemáticas sobre los datos de una columna específica de un DataFrame, en este caso, elevando al cuadrado los valores de la columna `Desfase`. Esto es útil para transformar los datos antes de su análisis o visualización.

````python
import numpy as np

# Elevar al cuadrado los valores de la columna 'Desfase'
desfase_cuadrado = np.square(datos_df['Desfase'].values)
print(f"Desfase al cuadrado: {desfase_cuadrado}")
````

## 4. Representación de datos

Para representar datos ya manipulados y obtener una gráfica con resultados, se pueden utilizar bibliotecas como `matplotlib`. Estas herramientas permiten crear visualizaciones claras y personalizables para analizar tendencias o patrones en los datos. Por ejemplo, podemos mostrar una gráfica de la `frecuencia`frente al `desfase_cuadrado`

````python
import matplotlib.pyplot as plt

# Crear objeto de figura
plt.figure()

# Definir error del 5% en la frecuencia
error = 0.05 * datos_df['Frecuencia'].values 

# Graficar los datos experimentales ya manipulados
plt.plot(desfase_cuadrado, datos_df['Frecuencia'] / 1e6, 'o-', label='Datos experimentales')

# Graficar la banda que representa el error
plt.fill_between(desfase_cuadrado, 
                 (datos_df['Frecuencia'] - error) / 1e6, 
                 (datos_df['Frecuencia'] + error) / 1e6, 
                 color='blue', alpha=0.3, label='Banda de error (5%)')

# Añadir etiquetas, título, leyenda y cuadrícula
plt.xlabel('Desfase al cuadrado (rad²) / Beta')
plt.ylabel('Frecuencia (MHz)')
plt.title('Relación entre Frecuencia y Desfase al cuadrado / Beta')
plt.grid(True)
plt.legend()

# Mostrar la gráfica
plt.show()
````
