### Frecuencia de corte
Para calcular la frecuencia de corte experimentalmente, se ha usado la relación:

$$
    \beta = w\sqrt{\mu \varepsilon}\sqrt{1- \omega_c^2 / \omega^2} ~~~~ \implies ~~~~ \beta^2 = \mu\varepsilon \omega^2 - \mu\varepsilon \omega_c^2
$$

que es válida para líneas de transmisión ideales. Puesto que el dieléctrico es aire, podemos tomar $\mu\varepsilon = \mu_0\varepsilon_0 = 1/c^2$. Representamos $\beta^2$ frente a $\omega^2$.

<img src="images/beta2_vs_w2.png" width=500> 


De forma teórica, la frecuencia de corte para el modo TE<sub>10</sub> debería ser:

$$
    \omega_c = \sqrt{\mu\varepsilon} ~~ k_c = \sqrt{\mu\varepsilon} ~~ \frac{\pi}{a}
$$

### Impedancias de carga
Para el cálculo de las impedancias de carga se ha usado:

$$
    |\Gamma| \, = \, \frac{SWR - 1}{SWR + 1}
$$ 

$$
    cos(\theta_{\Gamma} - 2 \beta z_{min}) = -1 \implies \theta_{\Gamma} = 2 \beta z_{min} - \pi - 2n\pi
$$

$$
    \Gamma = |\Gamma| e^{j \theta_{\Gamma}} = |\Gamma| e^{j (2 \beta z_{min} - \pi)}
$$

$$
    Z_L = Z_0 \, \frac{1 + \Gamma}{1 - \Gamma}
$$

## Conclusiones y análisis

¿Cuál es la frecuencia de corte teórica para el modo fundamental en la guía de ondas que has empleado?

La frecuencia de corte teórica calculada es de 6662.05 MHz (aprox. 6.66 GHz). Este valor proviene directamente del cálculo teórico usando la dimensión a = 2.25 cm de nuestra guía rectangular. El valor experimental real que obtenemos es de 6626 MHz. El valor teórico esperado (6662 MHz) y el valor experimental que hemos obtenido (6626 MHz) son prácticamente idénticos. Esta pequeña diferencia de apenas 36 MHz demuestra que las medidas tomadas con la sonda en el laboratorio han sido muy precisas y que el comportamiento de la guía real se ajusta casi a la perfección al modelo teórico ideal.

¿Cómo varía $\lambda_s$ con la frecuencia?

A medida que aumentamos la frecuencia de la señal, la longitud de onda dentro de la guía ($\lambda_s$) disminuye. Lo podemos comprobar directamente con nuestros datos experimentales: Para la frecuencia más baja que medimos (7586 MHz), la distancia entre mínimos nos dio una lambda_s grande, de unos 81.5 mm.Por el contrario, para la frecuencia más alta (9829 MHz), la $\lambda_s$ se redujo prácticamente a la mitad, midiendo unos 41.3 mm.

¿Cómo compara el valor obtenido para la longitud de onda en el sistema con la longitud de onda de propagación libre?

La longitud de onda dentro de la guía ($\lambda_s$) siempre es mayor que la del espacio libre ($\lambda_0$). Lo podemos comprobar numéricamente con nuestro primer dato. A la frecuencia de 9829 MHz, la longitud de onda en el aire es $\lambda_0$ = 30.5 mm. Sin embargo, nuestra medida en la guía nos dio un $\lambda_s$ = 41.2 mm. Queda demostrado que $\lambda_s$ > $\lambda_0$.

¿Es razonable el valor obtenido para $\mathcal{Z}_L$ en circuito abierto?

Respecto de la impedancia de carga en el caso de cortocircuito, vemos que la impedancia es bastante pequeña, como se esperaba. Sin embargo, en el caso del circuito abierto, se esperaba una impedancia infinita y sin embargo el resultado es de ($53 + j19.01 \pm 20 - j0.34$) $\Omega$, que está lejos de ser infinito. Esto es, en parte, porque es muy dificil replicar en un laboratorio las condiciones de circuito abierto, pues el propio aire y en general el ambiente puede afectar a las medidas. Por otro lado, vemos que tiene una componente compleja comparable con la parte real, cosa que no debería suceder al tener teóricamente un cortocircuito $\theta_\Gamma = 0$. Sin embargo, los errores que hemos obtenido de $\theta_\Gamma$ son muy grandes, comparables con $2\pi$, de forma que la fase del número complejo no nos da mucha información.
