# Práctica 2: Estudio Experimental de Guías de Ondas

En esta práctica de laboratorio se llevará a cabo la caracterización experimental de una guía de ondas rectangular. El objetivo principal es el estudio de la propagación de ondas electromagnéticas en guías de ondas a partir de medidas experimentales. Concretamente, se pretende determinar el diagrama de Brillouin del modo TE<sub>10</sub> (modo fundamental), que deberá evidenciar el comportamiento propio de un modo TE, así como determinar el valor de una determinada impedancia de carga desconocida a partir de la medida del parámetro SWR.

## Objetivos
- Determinar experimentalmente el diagrama de Brillouin para el modo fundamental de una guía de ondas rectangular de banda X.
- Estimar la frecuencia de corte del modo a partir del diagrama anterior y comparar con el resultado teórico.
- Determinar experimentalmente la impedancia de carga a partir de medidas de SWR y las posiciones de máximos de tensión en la guía desacoplada.

## Materiales
- Banco de guía de onda rectangular (banda X)
- Generador de microondas (banda X)
- Osciloscopio
- SWR meter
- Cavidad resonante calibrada en frecuencia con cristal detector.
- Sonda móvil de medida de tensión (cristal detector)
- Varias cargas (cortocircuito, circuito abierto, antena, terminación, lámina metálica)

## Resultados

- Diagrama de dispersión experimental, mostrando la dependencia de $\beta$ con la $\omega$ (diagrama de Brillouin).
- Estimación de la frecuencia de corte del modo TE<sub>10</sub> y comparación con el valor teórico esperado.
- Tabla de valores experimentales de: SWR y $\mathcal{Z}_L$; indicando sus respectivos errores.
