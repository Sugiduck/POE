# Práctica 3: Caracterización Experimental de Elementos de Circuitos

En esta práctica de laboratorio se llevará a cabo la caracterización experimental de un elemento de circuito de alta frecuencia. En concreto, se estudiará la potencia de salida por cada uno de sus puertos cuando se inyectan diferentes potencias por los demás. Para llevar a cabo las medidas de potencia se realizará, previamente, la calibración de un cristal detector. 

## Objetivos
- Calibrar un cristal detector para emplearlo en la medida de potencias de HF.
- Estudiar el comportamiento de un elemento de circuito con la potencia para inferir su funcionalidad. 

## Materiales
- Generador de H.F.
- Atenuador variable.
- Cristal detector.
- Medidor de potencia.
- Elementos de circuito.
- Adaptadores macho/hembra.

*Nota: los materiales de esta práctica cuentan todos con conexiones tipo N. (Salvo conexión banana al multímetro)*

## En el laboratorio
### Primera parte: calibración del cristal detector (tiempo estimado 1h)
Trabajar a una frecuencia de 5.5 GHz y una potencia de +10 dBm.

**IMPORTANTE: trabajar con el generador de microondas apagado mientras se montan/desmontan los montajes experimentales**

- Conectar el generador al atenuador variable mediante una línea coaxial.
- Conectar el cristal detector a la salida del atenuador variable y al voltímetro. 
- Para varias posiciones del atenuador medir la respuesta del cristal detector (tensión).
- Desconectar el cristal detector del atenuador variable y conectar el medidor de potencia. 
- Para las mismas posiciones del atenuador medir la potencia detectada. 

### Segunda parte: estudio experimental del comportamiento de un elemento de circuito (tiempo estimado 1h)
Trabajar a una frecuencia de 5.5 GHz y una potencia de +10 dBm. Elegir un elemento de circuito de tres puertos.

**IMPORTANTE: trabajar con el generador de microondas apagado mientras se montan/desmontan los montajes experimentales**

- Conectar el generador al atenuador variable mediante una línea coaxial.
- Conectar la salida del atenuador variable a uno de los puerto del elemento de circuito. 
- Conectar el cristal detector a otro puerto del elemento de circuito (y al voltímetro).
- Acoplar con una carga de 50 $\Omega$ el puerto libre del elemento de circuito.
- Para las mismas posiciones del atenuador medir la respuesta del cristal detector (tensión).
- Repetir las medidas anteriores para todas las combinaciones posibles en los puertos del elemento de circuito.

### Tercera parte: análisis de los datos experimentares y obtención de resultados (tiempo estimado 1h)
Al final de este notebook se muestra un ejemplo básico de cómo se podría emplear este mismo notebook de Jupyter (Python) como cuaderno de prácticas interactivo para tomar los datos experimentales, analizarlos/manipularlos y, finalmente, producir resultados (tablas y gráficas).

En la curva de calibración del cristal detector, observamos que para los valores más bajos de voltaje los puntos se acumulan en torno a una potencia máxima, por lo que no podremos usar esta curva de calibración para medir potencias mayores. Para valores de voltaje entre -1 mV y -12.5 mV se encuentra una zona en la que la función es monótona decreciente, por lo que sí permite determinar unívocamente la potencia en función del voltaje del cristal detector. Para voltajes cercanos a 0, lógicamente, la potencia también tiende a 0, de forma que la señal se hace indistinguible del ruido.
