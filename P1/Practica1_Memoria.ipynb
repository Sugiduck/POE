### Primera parte: Parámetros primarios

Tras realizar el procedimiento explicado en el apartado anterior, conseguimos medir los parámetros primarios de la línea para diferentes longitudes. Estos parámetros los podemos ver en la siguiente tabla:

<center>

|       $l$ (m)      |      $C^*$ (nF)  |$R^*$ ($\pm 0.1\,\Omega$)|
|:------------------:|:----------------:|:-----------------------:|
| $      100        $| $ 9.88 \pm 0.01$ | $         5.5         $ |
| $      200        $| $19.76 \pm 0.01$ | $        10.5         $ |

A partir de esta tabla, y usando que $v_f = 1/\sqrt{LC} \rightarrow L = 1/{C v_f^2}$, podemos obtener los valores de los parámetros primarios característicos de la línea haciendo una media ponderada de los valores correspondientes a cada longitud. De esta forma obtenemos:

$$ \boxed{C = (0.098816 \pm 0.000045)\,\mathrm{nF/m}}\,,\,\,\boxed{R = (0.05514 \pm 0.00027)\,\Omega\mathrm{/m}}\,,\,\,\boxed{L = (0.253 \pm 0.025)\,\mu\mathrm{H/m}}\,.$$

Para calcular la conductancia $G^*$ en el dieléctrico, medimos la resistencia en el mismo usando un multímetro con la línea en circuito abierto. Sin embargo, al ser dicha resistencia mucho mayor a la escala de medida que ofrece el multímetro ($R^*_G > 2$ $\text{G}\Omega$), lo único que conseguimos es estimar una cota superior para la conductancia en el dieléctrico $G^* < 0.5 \text{ nS}$. Como dicha cota superior es válida para cualquiera de las longitudes de línea que teníamos a nuestra disposición, podemos además dar la cota más pequeña posible para la conductancia caracerística tal que:

$$ G < 0.5/300 \text{ nS/m} \rightarrow \boxed{G < 1.667 \times 10^{-3} \text{ nS/m}}\,. $$



### Segunda parte: Constante de propagación

Utilizando una línea con $l = 100$ m, y acoplándola a una carga de $50\,\Omega$ (de forma que no haya onda reflejada que altere la medida), hicimos pasar varias señales armónicas de misma amplitud y con diferentes frecuencias para medir el desfase temporal y la diferencia de amplitudes entre la señal de entrada y la de salida. A partir de estas medidas podemos calcular la constante de propagación y de atencuación para cada una de las frecuencias de las señales empleadas de la siguiente manera:

- Sabiendo que en general, una señal armónica la podemos escribir como $A e^{-\alpha z} \cos(\omega t - \beta z + \delta)$, podemos ver que, sustituyendo en ambos extremos del cable, restando las fases resultantes y calculando el cociente entre las amplitudes, obtenemos:

    - Cociente de amplitudes: $V_e / V_s = A e^{-\alpha \cdot (0)} / A e^{-\alpha (l)} = e^{\alpha l} \rightarrow \boxed{\alpha = \frac{1}{l} \ln(V_e / V_s)} $
    - Desfase: $\Delta \phi = - \beta \cdot (0) + \beta \cdot (l) \rightarrow \boxed{\beta = \frac{\Delta \phi }{l}}$

- Con esto podemos calcular las constantes de atenuación y propagación para cada una de las señales de entrada, pudiendo representar así dichas constantes en función de la frecuencia.

- Además,  como la velocidad de fase es $v_f = \omega / \beta$, la pendiente resultante al representar $\omega$ vs. $\beta$ será precisamente la velocidad de fase característica de la línea, cosa que nos servirá para comparar con el valor dado en el protocolo $v_f=(2.0\pm0.1)\times 10^8\ (\mathrm{ms^{-1}})$.

- Hay que tener en cuenta que, a la hora de medir el desfase con el osciloscopio, mediremos desfase entre las crestas de dos oscilaciones que se muestren consecutivas en el osciloscopio. Si el desfase entre la señal de entrada y la de salida se hace mayor a $2\pi$, seguiremos midiendo valores de desfase entre $0$ y $2\pi$. Este hecho debemos corregirlo sumando un múltiplo de $2\pi$ que restaure la diferencia de fase total. Para ello, tomamos un valor de desfase de referencia y exigimos que el desfase no varíe bruscamente con la frecuencia. Queda abierta la pregunta de si el desfase de referencia verdaderamente está dentro del intervalo ($0$, $2\pi$)
