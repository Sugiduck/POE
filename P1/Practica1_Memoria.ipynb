# Práctica 1: Estudio Experimental de Líneas de Transmisión

En esta práctica de laboratorio se llevará a cabo la caracterización experimental de una línea de transmisión coaxial. El objetivo principal es determinar sus parámetros primarios ($R$, $L$, $G$, $C$) y la constante de propagación $\gamma = \alpha + j\ \beta$, a partir de medidas realizadas en el laboratorio. Finalmente, se representará el diagrama de Brillouin, que deberá evidenciar el comportamiento propio de un modo TEM (relación lineal entre $\beta$ y $\omega$).

## Objetivos
- Determinar los parámetros primarios: R, L, G y C.
- Realizar medidas experimentales de tensión y desfase en una línea coaxial.
- Determinar la constante de atenuación, $\alpha$, y la constante de fase, $\beta$; así como su dependencia con la frecuencia.
- Representar el diagrama de Brillouin y comprobar el comportamiento lineal del modo TEM.

## Materiales
- Generador de señales
- Líneas coaxiales (100 m)
- Osciloscopio
- Multímetro
- Cables y conectores

*Nota: los materiales de esta práctica cuentan todos con conexiones tipo BNC. (Salvo conexión banana al multímetro)*

## Métodos
### Primera parte: medida de parámetros primarios
Asumiendo los valores de impedancia característica de la línea $\mathcal{R}_0=50.0\pm 2.5\ (\Omega)$ y $v_f=(2.0\pm0.1)\times 10^8\ (\mathrm{ms^{-1}})$ (medidas tomadas en el experimento realizado en sesión de GG).

- Medir con la ayuda del multímetro la capacidad de la línea (línea en circuito abierto) con líneas de varias longitudes para determinar $C$.
- Medir con la ayuda del multímetro la resistencia de los conductores (línea cortocircuitada) con líneas de varias longitudes para determinar $R$.
- Medir con la ayuda del multímetro la resistencia del dieléctrico (línea en circuito abierto) con líneas de varias longitudes para determinar $G$.

### Segunda parte: medida de constante de propagación
Empleando señales armónicas de 10 V de tensión de pico a pico y 0 V de offset, y variando la frecuencia de 0.5 - 10 MHz, con la carga adaptada, se tomaron las siguientes medidas para las diferentes frecuencias:

- Amplitud de la señal a la entrada de la linea. 
- Amplitud de la señal a la salida de la linea. 
- Desfase entre las señales a la entrada y la salida.

Para calcular la conductancia $G$ en el dieléctrico, medimos la resistencia en el mismo usando un multímetro con la línea en circuito abierto. Sin embargo, al ser dicha resistencia mucho mayor a la escala de medida que ofrece el multímetro ($R_G > 2$ $\text{G}\Omega$), lo único que conseguimos es estimar una cota superior para la conductancia en el dieléctrico $\boxed{G < 0.5 \text{ nS}}$. 

### Segunda parte: Constante de propagación

Utilizando una línea con $l = 100$ m, y acoplándola a una carga de $50\,\Omega$ (de forma que no haya onda reflejada que altere la medida), hicimos pasar varias señales armónicas de misma amplitud y con diferentes frecuencias para medir el desfase temporal y la diferencia de amplitudes entre la señal de entrada y la de salida. A partir de estas medidas podemos calcular la constante de propagación y de atencuación para cada una de las frecuencias de las señales empleadas de la siguiente manera:

- Sabiendo que en general, una señal armónica la podemos escribir como $A e^{-\alpha z} \cos(\omega t - \beta z + \delta)$, podemos ver que, sustituyendo en ambos extremos del cable, restando las fases resultantes y calculando el cociente entre las amplitudes, obtenemos:

    - Cociente de amplitudes: $\Delta V = A e^{-\alpha \cdot (0)} / A e^{-\alpha (l)} = \alpha l \rightarrow \boxed{\alpha = \Delta V / l} $
    - Desfase: $\Delta \phi = - \beta \cdot (0) + \beta \cdot (l) \rightarrow \boxed{\beta = \Delta \phi / l}$

- Con esto podemos calcular las constantes de atenuación y propagación para cada una de las señales de entrada, pudiendo representar así dichas constantes en función de la frecuencia.

- Además,  como la velocidad de fase es $v_f = \omega / \beta$, la pendiente resultante al representar $\omega$ vs. $\beta$ será precisamente la velocidad de fase característica de la línea, cosa que nos servirá para comparar con el valor dado en el protocolo $v_f=(2.0\pm0.1)\times 10^8\ (\mathrm{ms^{-1}})$.
